```python
import pandas as pd

#lendo a base de dados
base_dados = pd.read_csv("cancelamentos.csv")

display(base_dados)
```

```python
#---------TRATAMENTO DE DADOS----------

# excluindo COLUNA (AXIS = 1) inútil para  a análise
base_dados = base_dados.drop("CustomerID", axis=1)
base_dados.info()
```

```python
#-------- ANÁLISE DE DADOS-------

#analisar a taxa de cancelamento da empresa
quantidade_cancelamento = base_dados["cancelou"].value_counts()
porcentagem_cancelamento = base_dados["cancelou"].value_counts(normalize=True).map("{:.2%}".format)

display(quantidade_cancelamento)
display(porcentagem_cancelamento)
```

```python
#analisar duração de contrato

duracao_contrato = base_dados["duracao_contrato"].value_counts()
porcentagem_contrato = base_dados["duracao_contrato"].value_counts("{:.2%}".format)

display (duracao_contrato)
display (porcentagem_contrato)

#agrupar por duração de contrado
grupos = base_dados.groupby("duracao_contrato")
#exibir a média das informações da tabela
display(grupos.mean(numeric_only=True))

#RESULTADO: todas as pessoas com assinaturas mensais cancelaram
```

```python
#continuando analise sem as pessoas de assinatura mensal

base_dados = base_dados[base_dados["duracao_contrato"] != "Monthly"]

display(base_dados.info())
display(base_dados["cancelou"].value_counts())
display(base_dados["cancelou"].value_counts(normalize=True).map("{:.2%}".format))

#RESULTADO: Agora tem  46% de cancelamento
```

```python
#analisando as assinaturas

assinaturas = base_dados["assinatura"].value_counts(normalize=True).map("{:.2%}".format)
agrupando_assinaturas_media = base_dados.groupby("assinatura").mean(numeric_only=True)

display(assinaturas)
display (agrupando_assinaturas_media)

#RESULTADO: valores muito próximos, nenhuma conclusão
```
